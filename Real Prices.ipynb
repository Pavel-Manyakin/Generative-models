```python
import yfinance as yf
import matplotlib.pyplot as plt

symbol = 'NFLX'

data = yf.download(symbol, start='2024-05-15', end='2024-05-16', interval='1m')
print(data)
if not data.empty:
    plt.figure(figsize=(14, 7))
    plt.plot(data.index, data['Close'], label='Price ')
    plt.title('Close price of Netflix on 15 may 2024')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid()
    plt.show()
else:
    print("No data for this date.")
```
